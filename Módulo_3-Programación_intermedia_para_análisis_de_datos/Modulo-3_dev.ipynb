# ¿Qué es el Análisis Exploratorio de los Datos?

`El Análisis Exploratorio de Datos (EDA)` es un proceso fundamental en el ámbito de la ciencia de datos, donde se aplican técnicas estadísticas y visuales para examinar y comprender la naturaleza de los datos. Este proceso no solo permite garantizar la objetividad e interoperabilidad de los datos, sino que también abre las puertas a un sinfín de descubrimientos valiosos.

El `EDA` se asemeja a una expedición a través de un territorio desconocido, donde cada paso nos acerca a comprender mejor los datos. A través de esta exploración, podemos:

- `Identificar y corregir errores:` Detectar posibles errores en la recolección o el ingreso de datos, asegurando la confiabilidad de la información.
- `Desenmascarar valores atípicos:` Descubrir valores anormales que podrían distorsionar los análisis, permitiendo una mejor interpretación de los datos.
- `Desvelar relaciones ocultas:` Explorar las correlaciones entre variables, revelando patrones y tendencias que podrían pasar desapercibidos.
- `Eliminar redundancias:` Identificar variables redundantes o irrelevantes, optimizando el análisis y obteniendo resultados más precisos.
- `Visualizar lo invisible:` Transformar datos complejos en gráficos claros y concisos, facilitando la comprensión y la comunicación de los hallazgos.

Recurso recomendado:
- [IBM ¿Análisis exploratorio de datos?](https://www.ibm.com/mx-es/topics/exploratory-data-analysis)

El `EDA` se utiliza en una amplia variedad de campos, incluyendo la ciencia, la ingeniería, la medicina, los negocios y las ciencias sociales. Es una herramienta esencial para cualquier científico de datos o analista que trabaje con conjuntos de datos.

Existen diversas técnicas para realizar un `EDA`, algunas de las más comunes son:

- `Descripción de las variables:` Se describe el tipo de variable (numérica, categórica), las medidas de resumen (media, mediana, moda, desviación estándar) y la distribución de la variable.
- `Visualización de datos:` Se utilizan gráficos y diagramas para visualizar la distribución de las variables, las relaciones entre ellas y la presencia de valores atípicos. Algunos ejemplos de gráficos comunes son histogramas, diagramas de dispersión, gráficos de barras y mapas de calor.
- `Análisis de relaciones:` Se exploran las relaciones entre las variables utilizando medidas de correlación y asociación.
Detección de anomalías: Se identifican valores atípicos o inusuales en el conjunto de datos.
- `Transformación de datos:` Se transforman las variables si es necesario para mejorar su distribución o facilitar el análisis.

`Beneficios del EDA:`

- `Mejora la comprensión de los datos:` El EDA permite a los analistas comprender mejor las características de los datos, lo que puede ser útil para identificar patrones importantes, detectar errores y tomar decisiones informadas.
- `Facilita el modelado de datos:` Un buen EDA puede ayudar a preparar los datos para el modelado de datos, lo que puede mejorar la precisión y el rendimiento de los modelos.
- `Ayuda a comunicar los resultados:` Los gráficos y diagramas generados durante el EDA pueden ser utilizados para comunicar los resultados de manera efectiva a otras personas.

# Diccionario de datos

En el mundo actual, donde los datos se han convertido en un activo crucial para las organizaciones, la gestión eficaz de la información es fundamental para el éxito. Un componente esencial de esta gestión es el diccionario de datos, una herramienta que proporciona una base de datos centralizada para almacenar y organizar información detallada sobre los elementos de datos utilizados en una organización.

`¿Qué es un diccionario de datos?`

Un diccionario de datos funciona como un repositorio centralizado que almacena información descriptiva sobre cada elemento de datos utilizado en un sistema o conjunto de datos. Esta información incluye:

- Nombre del elemento de dato: Un identificador único para cada elemento de datos.
- Descripción del elemento de dato: Una explicación detallada de lo que significa el elemento de datos y cómo se utiliza.
- Tipo de dato: El formato en que se almacena el dato (por ejemplo, número, texto, fecha).
- Longitud del dato: La cantidad máxima de caracteres o bytes que puede contener el dato.
- Origen del dato: De dónde proviene el dato (por ejemplo, una base de datos, un sensor, una entrada manual).
- Uso del dato: Cómo se utiliza el dato en el sistema (por ejemplo, para cálculos, informes, toma de decisiones).
- Reglas de negocio: Cualquier restricción o condición que se aplique al dato (por ejemplo, valores válidos, rangos permitidos).
- Propietario del dato: La persona o equipo responsable de la calidad y la integridad del dato.
- Beneficios del diccionario de datos:

`Un diccionario de datos bien mantenido ofrece numerosos beneficios para las organizaciones, incluyendo:`

- Mejora la comunicación y la colaboración: Al proporcionar una fuente única de información sobre los datos, los diccionarios de datos ayudan a que las personas que trabajan con los datos se entiendan mejor y colaboren de manera más eficaz.
- Garantiza la calidad de los datos: Los diccionarios de datos pueden ayudar a identificar y corregir errores en los datos, así como a garantizar que los datos sean consistentes y precisos.
- Facilita el uso de los datos: Los diccionarios de datos pueden hacer que los datos sean más fáciles de encontrar, comprender y utilizar para las personas que necesitan acceder a ellos.
- Cumplimiento de regulaciones: Los diccionarios de datos pueden ayudar a las organizaciones a cumplir con las regulaciones que requieren que documenten y protejan sus datos.

`¿Cómo se utiliza un diccionario de datos?`

Los diccionarios de datos se utilizan de diversas maneras, incluyendo:

- Documentación de nuevos conjuntos de datos: Cuando se crea un nuevo conjunto de datos, el diccionario de datos se puede utilizar para documentar la estructura, los elementos de datos y las relaciones del conjunto de datos.
- Búsqueda de información sobre elementos de datos: Los usuarios pueden consultar el diccionario de datos para encontrar información sobre un elemento de datos específico, como su definición, tipo de dato, origen y uso.
- Análisis de impacto de cambios: Cuando se realizan cambios en un sistema o conjunto de datos, el diccionario de datos se puede utilizar para identificar qué elementos de datos se verán afectados y cómo.
- Cumplimiento de regulaciones: El diccionario de datos se puede utilizar para demostrar a los auditores que la organización cumple con las regulaciones que requieren que documenten y protejan sus datos.

## Análisis univariable

El análisis univariable es una técnica estadística que se utiliza para estudiar y describir las características de una sola variable dentro de un conjunto de datos. Es un paso fundamental en el análisis exploratorio de datos y permite obtener una comprensión básica de la distribución, tendencia central, dispersión y valores atípicos de la variable en cuestión.

![Sin título.png](<attachment:Sin título.png>)

`Las principales técnicas de análisis univariable incluyen:`

- Medidas de tendencia central: Media, mediana, moda.
- Medidas de dispersión: Rango, desviación estándar, varianza.
- Análisis de frecuencias: Tablas de frecuencias, gráficos de barras o pie.
- Representaciones gráficas: Histogramas, diagramas de cajas, diagramas de dispersión.

`El análisis univariable es útil para:`

- Identificar la estructura básica de la variable.
- Resumir las características principales de la variable.
- Explorar la distribución de la variable.
- Detectar valores atípicos o inusuales.
- Preparar el terreno para análisis multivariantes más complejos.

```python
# Calcular los 5 departamentos con puntajes globales más altos
top5_departamentos = df_acumulado_saber_11.groupby('ESTU_DEPTO_RESIDE')['PUNT_GLOBAL'].mean().nlargest(5).index

# Filtrar los datos para incluir solo los departamentos del top 5
datos_top5_departamentos = df_acumulado_saber_11[df_acumulado_saber_11['ESTU_DEPTO_RESIDE'].isin(top5_departamentos)]

# Agrupar datos por periodo y departamento de residencia
punt_global_por_periodo_depto = datos_top5_departamentos.groupby(['PERIODO', 'ESTU_DEPTO_RESIDE'])['PUNT_GLOBAL'].mean().reset_index()

# Crear la línea de tiempo
plt.figure(figsize=(14, 8))

# Usar una paleta de colores más atractiva
palette = sns.color_palette("coolwarm", len(top5_departamentos))

# Crear la línea de tiempo con puntos y etiquetas de valor
sns.lineplot(x='PERIODO', y='PUNT_GLOBAL', hue='ESTU_DEPTO_RESIDE', data=punt_global_por_periodo_depto, palette=palette, marker='o')

# Añadir etiquetas de valor a cada punto
for line in range(0, punt_global_por_periodo_depto.shape[0]):
    plt.text(punt_global_por_periodo_depto.PERIODO[line], punt_global_por_periodo_depto.PUNT_GLOBAL[line],
             round(punt_global_por_periodo_depto.PUNT_GLOBAL[line], 2), horizontalalignment='left', size='small', color='black')

# Configurar el gráfico
plt.title('Línea de tiempo de PUNT_GLOBAL por PERIODO en función de ESTU_DEPTO_RESIDE (Top 5 departamentos)')
plt.xlabel('Periodo')
plt.ylabel('Puntaje Global Promedio')

# Añadir una grilla de fondo para facilitar la lectura
plt.grid(True, which='both', linestyle='--', alpha=0.5)

# Configurar la leyenda para mejorar la presentación
plt.legend(title='Departamento de Residencia')

# Mostrar el gráfico
plt.show()
```

### Análisis de distribución de variables cuantitativas

Visualización de la distribución de variables numéricas:

- Objetivo: Utilizar histogramas, diagramas de densidad y boxplots para visualizar la distribución de las variables numéricas (como PUNT_GLOBAL, PUNT_LECTURA_CRITICA, PUNT_MATEMATICAS, etc.).

- Aplicación: Esto te ayudará a identificar posibles valores atípicos y patrones en la distribución de los datos.

```python
# Función para visualizar la distribución de variables numéricas
def visualizar_distribucion_variables_numericas(df, variables_numericas):
    # Crear una figura con una cuadrícula de 3 columnas y 2 filas
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    axes = axes.flatten()  # Aplanar la cuadrícula de ejes para iterar fácilmente
    
    # Iterar sobre cada variable numérica y el correspondiente eje
    for i, variable in enumerate(variables_numericas):
        # Crear un histograma con una curva de densidad superpuesta para cada variable numérica
        ax = axes[i]
        sns.histplot(df[variable], kde=True, color='blue', bins=30, ax=ax)
        
        # Añadir etiquetas de frecuencia en las barras
        for patch in ax.patches:
            height = patch.get_height()
            ax.text(patch.get_x() + patch.get_width() / 2, height, f'{height:.0f}', ha='center', va='bottom', fontsize=8)
        
        # Configurar el título y etiquetas
        ax.set_title(f'Distribución de la variable {variable}')
        ax.set_xlabel(variable)
        ax.set_ylabel('Frecuencia')
    
    # Eliminar los ejes sobrantes si hay menos variables que subtramas
    for j in range(len(variables_numericas), len(axes)):
        fig.delaxes(axes[j])
    
    # Ajustar el diseño y mostrar la figura
    plt.tight_layout()
    plt.show()

# Lista de variables numéricas para analizar
variables_numericas = ['PUNT_GLOBAL', 'PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES']

# Llamar a la función para visualizar la distribución de las variables numéricas
visualizar_distribucion_variables_numericas(df_acumulado_saber_11, variables_numericas)
```
